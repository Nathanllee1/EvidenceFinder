# Imports

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

# Open Page

```python
page = requests.get("https://www.litcharts.com/")
books = BeautifulSoup(page.content, 'html.parser')
```

# Scraper

```python
book_link = []
quotetheme = {}
book_counter = 1

#runningtime = time.localtime(time.time())

#Find books
for book in books.find_all(class_='book hoverable', href=True):
    
    #make into a real url
    book_counter += 1

    book_url = 'https://www.litcharts.com' + book['href'] + '/quotes'
    book_link.append(book_url)
    
    #Open quotepage
    for links in book_link:
        quote_page_request = requests.get(str(links))
        quote_page = BeautifulSoup(quote_page_request.content, 'html.parser')
        
        quote_counter = 0

        #Find quotes and themes
        for quoteblock in quote_page.find_all(class_='summary-section list-group'):
            if quoteblock.find(class_='quote-text') and quoteblock.find(class_='list-group-item quote') is not None:
                quote_ = quoteblock.find(class_='quote-text')
                quote = str(quote_.get_text())
                #print(quote)

                theme = quoteblock.find(class_='list-group-item quote').attrs['data-themes']
                quotetheme.update({quote : theme})
                quote_counter += 1
                #print(theme)
        '''
        for otherquoteblock in quote_page.find_all(class_='list-group-item quote'):
            if otherquoteblock.find(class_='quote-text') and quoteblock.find(class_='list-group-item quote') is not None:
                quote_ = quoteblock.find(class_='quote-text')
                quote = str(quote_.get_text())
                #print(quote)

                theme = quoteblock.find(class_='list-group-item quote').attrs['data-themes']
                quotetheme.update({quote : theme})
                quote_counter += 1
                #print(theme)
    '''
    actualquotecount = len(quotetheme)
    print('Found: ' + book_url + '   book number ' + str(book_counter) + ' / 791   found ' + str(quote_counter) + ' quotes   Total quotes: ' + str(actualquotecount))
    quote_counter = 0
    if book_counter == 21:
        break
```

```python
dataframe = pd.DataFrame.from_dict(quotetheme, orient='index')
#dataframe = pd.DataFrame.from_dict(list(quotetheme.items()), columns = ['Quotes', 'Themes'], orient='index')

dataframe.to_csv('quotes' + str(book_counter) + '.csv', sep='\t')
dataframe
```
