# Generate False Data

```python
import random

df = pd.read_csv('scrapeddata.csv')
falsedata = []
#compare two lists
def compare(one, two):
    two = set(two)
    one = set(one)
    if len(one.intersection(two)) > 0: 
        return(True)  
    return(True)

#select a random theme
def randomtheme():
    randomnum = random.randint(1, df.shape[0]) - 1
    randomtheme = df.iloc[randomnum, 2]
    #print(randomtheme)
    return randomtheme

#append a new frame
def appendframe(quote, randtheme):
    newdict = {'Quotes' : quote, 'Themes' : randtheme, 'Boolean' : 'no'}
    falsedata.append(newdict)    
    
rowcounter = 0
for row in df.iterrows():
    print(rowcounter, end=' ')
    #select data row by row
    selecteddata = df.iloc[rowcounter,:]
    
    #get quote
    samequote = selecteddata['Quotes']
    sametheme = selecteddata['Themes']
    
    #search for different data  
    
    randtheme = randomtheme()
    while compare(sametheme, randtheme) == False:
        randtheme = randomtheme()
        compare(sametheme, randtheme)
        
    appendframe(samequote, randtheme)
        
    #somebody help with this
    '''
    else:
        #keep trying until a non match is found
        while compare(sametheme, randomtheme_) == False:
            randtheme = randomtheme()
            compare(sametheme, randomtheme_)
            
        appendframe(samequote, randtheme)
        
    if compare(sametheme, randomtheme_) == True:
        appendframe(samequote, randtheme)
        print('Yeehaw' + str(rowcounter), end=' ')
        '''
    rowcounter += 1
```

# Change 1 or 0 to yes and no

```python
df = pd.read_csv('scrapeddata.csv')

#change 1 to yes
df['Boolean'] = 'yes'

falseframe = pd.DataFrame(falsedata)

#change 0 to no
falseframe['Boolean'] = 'no'

final = pd.concat([falseframe, df])
print(final)
```
