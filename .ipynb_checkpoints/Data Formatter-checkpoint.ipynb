# Generate False Data

```python
import random


falsedata = []
#compare two lists
def compare(one, two):
    two = set(two)
    one = set(one)
    if len(one.intersection(two)) > 0: 
        return(False)  
    return(True)

#select a random theme
def randomtheme():
    randomnum = random.randint(1, df.shape[0])
    randomtheme = df.iloc[randomnum, 2]
    #print(randomtheme)
    return randomtheme

#append a new frame
def appendframe(quote, randtheme):
    newdict = {'Quotes' : quote, 'Themes' : randtheme, 'Boolean' : 0}
    falsedata.append(newdict)    
    
rowcounter = 0
for row in df.iterrows():
    #select data row by row
    selecteddata = df.iloc[rowcounter,:]
    
    #get quote
    samequote = selecteddata['Quotes']
    sametheme = selecteddata['Themes']
    
    #search for different data  
    randomtheme_ =  randomtheme()
    if compare(sametheme, randomtheme_) == True:
        randtheme = randomtheme()
        appendframe(samequote, randtheme)
    else:
        #keep trying until a non match is found
        while compare(sametheme, randomtheme_) == False:
            randtheme = randomtheme()
            compare(sametheme, randomtheme_)
            
        appendframe(samequote, randtheme)
        
    rowcounter += 1
```

```python
falseframe = pd.DataFrame(falsedata)
df.append(falseframe, ignoreindex = True)
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = df['column name']
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names())
print(X.shape)
```
