# Imports

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# Open Page

```python
page = requests.get("https://www.litcharts.com/")
books = BeautifulSoup(page.content, 'html.parser')
```

# Scraper

```python
book_counter = 0
listofdicts = []

#Find books
for book in books.find_all(class_='book hoverable', href=True):
    
    #make into a real url
    book_counter += 1

    book_url = 'https://www.litcharts.com' + book['href'] + '/quotes'
    
    #Open quotepage
    
    quote_page_request = requests.get(str(book_url))
    quote_page = BeautifulSoup(quote_page_request.content, 'html.parser')

    quote_counter = 0

    #Find quotes and themes
    for quoteblock in quote_page.find_all(class_='summary-section list-group'):
        if quoteblock.find(class_='quote-text') and quoteblock.find(class_='list-group-item quote') is not None:
            quote_ = quoteblock.find(class_='quote-text')
            quote = str(quote_.get_text())
            #print(quote)

            theme = quoteblock.find(class_='list-group-item quote').attrs['data-themes']
            #quotetheme.update({quote : theme})

            #make and append a row to dataframe
            dictionary = {'Quotes': quote, 'Themes': theme, 'Boolean': 1}
            listofdicts.append(dictionary)

            quote_counter += 1
            #print(theme)

    print('Found: ' + book_url + '   book number ' + str(book_counter) + ' / 791   found ' + str(quote_counter) + ' quotes')
    quote_counter = 0
    if book_counter == 15:
        break
```

# Generate Fake Data

# Save to dataframe and CSV

```python
df = pd.DataFrame(listofdicts)
csvname = 'quotes' + str(df.shape) + '.csv'
print(csvname)
df.to_csv(csvname, sep='\t')

pd.set_option('display.width', 4000)
df
```

```python
#compare two lists
def compare(one, two):
    two = set(two)
    one = set(one)
    if len(one.intersection(two)) > 0: 
        return(False)  
    return(True)

#select a random theme
def randomtheme():
    randomtheme = df.sample(n=1)
    randomtheme = randomtheme['Themes']
    #print(randomtheme)
    return randomtheme

#append a new frame
def appendframe(quote, randtheme):
    newdict = {'Quotes' : quote, 'Themes' : randtheme, 'Boolean' : 0}
    listofdicts.append(newdict)    
    
rowcounter = 0
for row in df.iterrows():
    #select data row by row
    selecteddata = df.iloc[rowcounter,:]
    
    #get quote
    samequote = selecteddata['Quotes']
    
    sametheme = selecteddata['Themes']
    
    #search for different data  
    randomtheme_ =  randomtheme()
    if compare(sametheme, randomtheme_) == True:
        randtheme = randomtheme()
        appendframe(samequote, randtheme)
    else:
        randtheme = randomtheme()
        appendframe(samequote, randtheme)
        
    rowcounter += 1
```
